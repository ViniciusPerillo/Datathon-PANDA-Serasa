```python
def generate_combinations(name):
    words = name.split()
    combinations = []
    for i in range(1, len(words) + 1):
        combinations.extend([' '.join(words[:i]), ''.join([word[0] for word in words[:i]])])
    return combinations

def process_dataframe(df):
    # Assuming 'Name' is the column containing names
    name_column = 'NOME_MUNICIPIO'

    # Generate initials and combinations
    df['Initials'] = df[name_column].apply(lambda x: ''.join(word[0] for word in x.split()))
    df['Combinations'] = df[name_column].apply(generate_combinations)

    # Expand the combinations list into separate columns
    df = pd.concat([df, df['Combinations'].apply(pd.Series).add_prefix('Combination_')], axis=1)

    # Drop the original 'Combinations' column
    df = df.drop(columns=['Combinations'])

    return df
```

```python
total_cidades = len(df)

# Contagem de cidades com HTML
cidades_com_html = df[df['HTML'] != '']
cidades_com_html_count = len(cidades_com_html)

# Calcular a porcentagem de cidades com HTML em relação ao total de cidades
porcentagem_com_html = (cidades_com_html_count / total_cidades) * 100

# Criar um DataFrame para o gráfico
dados_grafico = pd.DataFrame({
    'Categoria': ['Coletada', 'Não Coletada'],
    'Porcentagem': [porcentagem_com_html, 100 - porcentagem_com_html]
})

plt.figure(figsize=(8, 6))
plt.pie(dados_grafico['Porcentagem'], labels=dados_grafico['Categoria'], autopct='%1.1f%%', startangle=140,
        colors=['skyblue', 'lightgreen'])
plt.title('')
plt.axis('equal')
plt.show()
```

```python
cidades_com_html = df[df['HTML'] != '']

# Contar o número de cidades com HTML por estado
contagem_por_estado = cidades_com_html.groupby('UF').size().reset_index(name='Contagem')

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(contagem_por_estado['UF'], contagem_por_estado['Contagem'], color='skyblue')
plt.title('')
plt.xlabel('Estados')
plt.ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
html_por_cidade = df[df['HTML'] != ''].groupby('NOME_MUNICIPIO')['HTML'].count()

# Calcular a média de HTMLs por cidade
media_html_por_cidade = html_por_cidade.mean()

# Criar um gráfico de barras com a média de HTMLs por cidade
plt.figure(figsize=(8, 6))
html_por_cidade.plot(kind='bar', color='skyblue')
plt.axhline(y=media_html_por_cidade, color='red', linestyle='--', label=f'Média: {media_html_por_cidade:.2f}')
plt.title('')
plt.xlabel('Cidades')
plt.ylabel('')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
