```python

from bs4 import BeautifulSoup
import requests as requests
import re
import pandas as pd


# URL do site
url = 'https://www.lajeado.to.gov.br'
response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')

links = soup.find_all('a', href=re.compile(r'^(?!.*noticia).*secretaria', flags=re.IGNORECASE))

#caso esteja em orgaos
if not links:
  links = soup.find_all('a', href=re.compile(r'^(?!.*noticia).*orgaos', flags=re.IGNORECASE))

#caso nao tenha secretaria diretamente associado ao link
if not links:
 links = soup.find_all('a')
 links = [link['href'] for link in links if 'secretaria' in link.get_text().lower()]

#caso nao tenha orgaos diretamente associado ao link
if not links:
 links = soup.find_all('a')
 links = [link['href'] for link in links if 'orgaos' in link.get_text().lower()]


for link in links:
    link = link["href"]
    print(f'Link: {link}')
```
